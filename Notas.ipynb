```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


df = pd.read_csv("notas_estudiantes.csv")

notas_asignaturas = df.groupby("asignatura")["nota"].median()

print(notas_asignaturas)


```

```python
notas_asignaturas.plot(kind="bar")
plt.title("Notas Medias por Asignatura")
plt.xlabel("Asignatura")
plt.ylabel("Nota")
plt.show()
```

```python
estudiante = "Estudiante_17"
df_estudiante = df[df["estudiante"] == estudiante]

plt.plot(df_estudiante["asignatura"], df_estudiante["nota"], marker="o")
plt.title(f"Notas de {estudiante}")
plt.xlabel("Asignatura")
plt.ylabel("Nota")
plt.ylim(0, 10)
for i, nota in enumerate(df_estudiante["nota"]):
    plt.text(i, nota, str(nota), ha="center", va="bottom")
plt.show()
```
